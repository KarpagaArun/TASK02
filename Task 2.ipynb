```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "menu.csv"  # Update this path if necessary
df = pd.read_csv(file_path)
```

```python
print("Dataset Overview:\n")
```

```python
print(df.info())
```

```python
print("\nFirst 5 rows:\n", df.head())
```

```python
print("\nSummary Statistics:\n", df.describe())
```

```python
print("\nMissing Values:\n", df.isnull().sum())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Calories'], bins=30, kde=True, color='blue')
plt.title('Distribution of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(12, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(12, 5))
category_calories = df.groupby("Category")['Calories'].mean().sort_values()
sns.barplot(x=category_calories.index, y=category_calories.values, palette='viridis')
plt.xticks(rotation=90)
plt.title('Average Calories per Category')
plt.xlabel('Category')
plt.ylabel('Average Calories')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Total Fat'], y=df['Calories'], hue=df['Category'], alpha=0.7)
plt.title('Total Fat vs Calories')
plt.xlabel('Total Fat (g)')
plt.ylabel('Calories')
plt.show()
```

```python
df.to_csv("cleaned_menu.csv", index=False)
```
